```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

requests = []
responses = []

from seleniumwire import webdriver
from selenium.webdriver.chrome.options import Options
import time

headless = False  
chrome_options = Options()

if headless:
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--disable-gpu")

chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("--start-maximized")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
driver.execute_cdp_cmd("Network.setBlockedURLs", {"urls": ["*.png", "*.jpg", "*.jpeg", "*.gif", "*.woff", "*.ttf"]})
driver.execute_cdp_cmd("Network.enable", {})

try:
    for match in finals_sofascore:
        match_id = match["id"]
        match_title = match["title"]
        match_url = f"https://www.sofascore.com/fr/football/match/{match_title}/{match_id}"

        print(f"Processing match: {match_title} (ID: {match_id})")
        driver.get(match_url)

        print(f"URL: {match_url} loaded.")

        for request in driver.requests:
            if request.response:
                print(f"Request URL: {request.url}")
                requests.append(request.url)
                responses.append(request.response.status_code)


finally:
    driver.quit()
```

```python
import asyncio
from playwright.async_api import async_playwright

requests = []
responses = []

headless = False  # keep the same as before

async with async_playwright() as p:
    browser = await p.chromium.launch(headless=headless)
    context = await browser.new_context()
    page = await context.new_page()

    for match in finals_sofascore:
        match_id = match["id"]
        match_title = match["title"]
        match_url = f"https://www.sofascore.com/fr/football/match/{match_title}/{match_id}"

        print(f"Processing match: {match_title} (ID: {match_id})")

        match_requests = []
        match_responses = []

        # Intercept requests and responses
        page.on("request", lambda request: match_requests.append(request.url))
        page.on("response", lambda response: match_responses.append(response.status))

        # Navigate to page
        await page.goto(match_url, timeout=100000)
        await page.wait_for_timeout(10000)  # wait 10s

        # Save captured requests/responses
        requests.extend(match_requests)
        responses.extend(match_responses)

        print(f"Captured {len(match_requests)} requests for {match_title}")

    await browser.close()

# Run the async function in a notebook or async loop
# Example:
# asyncio.run(process_matches(finals_sofascore))
```

```python
results = []
async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context()
        results = []

        for url in list(set([x for x in requests if x.startswith("https://www.sofascore.com/api/v1")])):
            try:
                # Use Playwright's request context to fetch API directly
                response = await context.request.get(url)
                if response.ok:
                    data = await response.json()
                    results.append({"url": url, "json": data})
                    print(f"Fetched JSON from {url}")
                else:
                    print(f"Failed {url} - status: {response.status}")
            except Exception as e:
                print(f"Error fetching {url}: {e}")

        await browser.close()
```

```python
import json
df = pd.DataFrame([{"url": r["url"], "json_text": str(r["json"])} for r in results])
df.to_csv("sofascore_api_data.csv", index=False)

# Save results to a proper JSON file
with open("sofascore_api_data.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)
```
