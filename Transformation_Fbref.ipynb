```python
# Load and display the saved CSV file
import pandas as pd

csv_file = "fbref_matches_full_html.csv"
data = pd.read_csv(csv_file)

print(data.head())
```

```python
from bs4 import BeautifulSoup, Comment

# assuming your dataframe is called df
html_first_match = data.loc[0, "html"]

soup = BeautifulSoup(html_first_match, "lxml")  # or "html.parser"
```

```python
from io import StringIO

tables = []

for table in soup.find_all("table"):
    table_id = table.get("id")
    html_io = StringIO(str(table))
    df_table = pd.read_html(html_io)[0]
    tables.append((table_id, df_table))
```

```python
for comment in soup.find_all(string=lambda text: isinstance(text, Comment)):
    if "<table" in comment:
        comment_soup = BeautifulSoup(comment, "lxml")
        for table in comment_soup.find_all("table"):
            table_id = table.get("id")
            html_io = StringIO(str(table))
            df_table = pd.read_html(html_io)[0]
            tables.append((table_id, df_table))
```

```python
tables_dict = {}

for table_id, df_table in tables:
    if table_id is None:
        table_id = f"table_{len(tables_dict)}"
    tables_dict[table_id] = df_table
```

```python
import pandas as pd
from bs4 import BeautifulSoup

html = data.loc[0, "html"]
soup = BeautifulSoup(html, "lxml")

teams_data = []

for team_div in soup.find_all("div", class_="scorebox_team"):
    team_id = team_div.get("id")

    # Team name & URL
    team_link = team_div.find("strong").find("a")
    team_name = team_link.text.strip()
    team_url = team_link.get("href")

    # Logo
    logo_img = team_div.find("img", class_="teamlogo")
    logo_url = logo_img.get("src") if logo_img else None

    # Score & xG
    score_div = team_div.find("div", class_="score")
    xg_div = team_div.find("div", class_="score_xg")

    goals = int(score_div.text) if score_div else None
    xg = float(xg_div.text) if xg_div else None

    # Manager
    manager = None
    captain = None

    for dp in team_div.find_all("div", class_="datapoint"):
        label = dp.find("strong").text.strip()
        value = dp.text.replace(label, "").replace(":", "").strip()

        if label == "Manager":
            manager = value
        elif label == "Captain":
            captain = value

    teams_data.append({
        "team_id": team_id,
        "team_name": team_name,
        "team_url": team_url,
        "logo_url": logo_url,
        "goals": goals,
        "xg": xg,
        "manager": manager,
        "captain": captain
    })

df_scorebox_team = pd.DataFrame(teams_data)
```

```python
import pandas as pd
from bs4 import BeautifulSoup

html = data.loc[0, "html"]
soup = BeautifulSoup(html, "lxml")

lineup_rows = []

for lineup_div in soup.find_all("div", class_="lineup"):
    table = lineup_div.find("table")
    rows = table.find_all("tr")

    team_name = None
    formation = None
    is_bench = False

    for row in rows:
        # Team header
        th = row.find("th")
        if th:
            text = th.text.strip()
            if "Bench" in text:
                is_bench = True
            else:
                team_name, formation = text.split(" (")
                formation = formation.replace(")", "")
                is_bench = False
            continue

        cols = row.find_all("td")
        if len(cols) != 2:
            continue

        number = cols[0].text.strip()
        player_link = cols[1].find("a")

        player_name = player_link.text.strip()
        player_url = player_link.get("href")

        icons = cols[1].find_all("div", class_="event_icon")

        lineup_rows.append({
            "team_name": team_name,
            "formation": formation,
            "player_number": int(number),
            "player_name": player_name,
            "player_url": player_url,
            "is_bench": is_bench,
            "yellow_card": any("yellow_card" in i["class"] for i in icons),
            "substituted_in": any("substitute_in" in i["class"] for i in icons),
        })

df_lineup = pd.DataFrame(lineup_rows)
```

```python
extra_stats = []

extra_div = soup.find("div", id="team_stats_extra")

for block in extra_div.find_all("div", recursive=False):
    headers = block.find_all("div", class_="th")
    team_left = headers[0].text.strip()
    team_right = headers[2].text.strip()

    stats = block.find_all("div")[3:]

    for i in range(0, len(stats), 3):
        left_value = stats[i].text.strip()
        stat_name = stats[i + 1].text.strip()
        right_value = stats[i + 2].text.strip()

        extra_stats.append({
            "team": team_left,
            "stat": stat_name,
            "value": int(left_value)
        })

        extra_stats.append({
            "team": team_right,
            "stat": stat_name,
            "value": int(right_value)
        })

df_team_stats_extra = pd.DataFrame(extra_stats)
```

```python
events = []

for event_div in soup.find_all("div", class_="event"):
    minute_div = event_div.find("div")
    minute = minute_div.text.strip().replace("’", "") if minute_div else None

    icon = event_div.find("div", class_="event_icon")
    event_type = icon["class"][1] if icon else None

    player_link = event_div.find("a")
    player_name = player_link.text.strip() if player_link else None
    player_url = player_link.get("href") if player_link else None

    logo = event_div.find("img", class_="teamlogo")
    team_logo = logo.get("src") if logo else None

    events.append({
        "minute": minute,
        "event_type": event_type,
        "player_name": player_name,
        "player_url": player_url,
        "team_logo": team_logo
    })

df_events = pd.DataFrame(events)
```
